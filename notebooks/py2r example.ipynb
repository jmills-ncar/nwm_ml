```python
from sklearn.datasets import load_iris
import pandas as pd
%load_ext rpy2.ipython
%matplotlib inline
```

```python
iris = load_iris()
iris = pd.DataFrame(iris.data,columns=iris.feature_names)
iris.head()
```

# Python

```python
import seaborn as sns
sns.lmplot(x='sepal width (cm)',y='sepal length (cm)',data=iris)
```

# R

```python
%%R -i iris
plot(iris$sepal.width..cm.,iris$sepal.length..cm.)
```
